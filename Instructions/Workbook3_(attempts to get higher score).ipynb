```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import tensorflow as tf

#  Import and read the charity_data.csv.
import pandas as pd 
application_df = pd.read_csv("Resources/charity_data.csv")
application_df.head()
```

```python
# Drop the non-beneficial ID columns, 'EIN' and 'NAME'.
#  YOUR CODE GOES HERE
application_df = application_df.drop(['EIN', 'NAME'], axis=1)
application_df
```

```python
application_dict = dict(application_df['APPLICATION_TYPE'].value_counts())
# Choose a cutoff value and create a list of application types to be replaced
# use the variable name `application_types_to_replace`
#  YOUR CODE GOES HERE

# cutoff point: any application that has less than 500 counts
application_types_to_replace = []
for i in application_dict:
    if application_dict[i] < 100:
        application_types_to_replace.append(i)


# Replace in dataframe
for app in application_types_to_replace:
    application_df['APPLICATION_TYPE'] = application_df['APPLICATION_TYPE'].replace(app,"Other")

# Check to make sure binning was successful
application_df['APPLICATION_TYPE'].value_counts()
```

```python
classification_dict = dict(application_df['CLASSIFICATION'].value_counts())

# Choose a cutoff value and create a list of classifications to be replaced
# use the variable name `classifications_to_replace`
#  YOUR CODE GOES HERE
classifications_to_replace = []
for i in classification_dict:
    if classification_dict[i] < 100:
        classifications_to_replace.append(i)

# Replace in dataframe
for cls in classifications_to_replace:
    application_df['CLASSIFICATION'] = application_df['CLASSIFICATION'].replace(cls,"Other")
    
# Check to make sure binning was successful
application_df['CLASSIFICATION'].value_counts()
```

```python
application_df['AFFILIATION'].value_counts()
```

```python
# drop all inactive apps: STATUS == 0
application_df = application_df.drop(application_df.loc[application_df['STATUS'] == 0].index, axis=0)

# drop STATUS column, only consider the active applications
application_df = application_df.drop(columns='STATUS')
```

```python
# get dummies features
dummies_application_df = pd.get_dummies(application_df)
```

```python
application_df['INCOME_AMT'].value_counts()
```

```python
# Split our preprocessed data into our features and target arrays
#  YOUR CODE GOES HERE
y = dummies_application_df['IS_SUCCESSFUL']
X = dummies_application_df.drop(columns='IS_SUCCESSFUL')

# Split the preprocessed data into a training and testing dataset
#  YOUR CODE GOES HERE
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=78)
```

```python
# Create a StandardScaler instances
scaler = StandardScaler()

# Fit the StandardScaler
X_scaler = scaler.fit(X_train)

# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
X_train_scaled.shape
```

```python
# Define the model - deep neural net, i.e., the number of input features and hidden nodes for each layer.
#  YOUR CODE GOES HERE

nn_model = tf.keras.models.Sequential()

# First hidden layer
#  YOUR CODE GOES HERE
nn_model.add(tf.keras.layers.Dense(units=30, activation="relu", input_dim=49))

# Second hidden layer
#  YOUR CODE GOES HERE
nn_model.add(tf.keras.layers.Dense(units=8, activation="sigmoid"))

# Output layer
#  YOUR CODE GOES HERE
nn_model.add(tf.keras.layers.Dense(units=1, activation="sigmoid"))
# Check the structure of the model
nn_model.summary()
```

```python
nn_model.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"])

# Fit the model to the training data
fit_model = nn_model.fit(X_train_scaled, y_train, epochs=100)
```

```python
# MODEL 2: with 3 layers

nn_model_2 = tf.keras.models.Sequential()

# First hidden layer
#  YOUR CODE GOES HERE
nn_model_2.add(tf.keras.layers.Dense(units=20, activation="relu", input_dim=49))

# Second hidden layer
#  YOUR CODE GOES HERE
nn_model_2.add(tf.keras.layers.Dense(units=15, activation="sigmoid"))

# third layer
nn_model_2.add(tf.keras.layers.Dense(units=10, activation="sigmoid"))


# Output layer
#  YOUR CODE GOES HERE
nn_model_2.add(tf.keras.layers.Dense(units=1, activation="sigmoid"))
# Check the structure of the model
nn_model_2.summary()
```

```python
nn_model_2.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"])

# Fit the model to the training data
fit_model_2 = nn_model_2.fit(X_train_scaled, y_train, epochs=100)
```

```python
# Evaluate the model 1 with 2 layers using the test data
model_loss, model_accuracy = nn_model.evaluate(X_test_scaled,y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```

```python
# Evaluate the model 2 with 3 layers using the test data
model_loss_2, model_accuracy_2 = nn_model_2.evaluate(X_test_scaled,y_test,verbose=2)
print(f"Loss: {model_loss_2}, Accuracy: {model_accuracy_2}")
```
